```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rcParams["figure.figsize"] = [30, 15]
import seaborn as sns
sns.set_theme(color_codes=True)
params = {'legend.fontsize':40,
         'axes.labelsize': 35,
         'axes.titlesize':'x-large',
         'xtick.labelsize':40,
         'ytick.labelsize':40,
         'lines.linewidth':6,
        'axes.xmargin':0}
mpl.rcParams.update(params)
```

```python
#Styling for Box Plots
boxprops = dict(linestyle='-', linewidth=4, color='g')
medianprops = dict(linestyle='-', linewidth=4, color='r')
flierprops=dict(linestyle='-', linewidth=2)
whiskerprops=dict(linestyle='-', linewidth=2)
capprops=dict(linestyle='-', linewidth=2)
meanpointprops = dict(marker='D', markeredgecolor='black',
                      markerfacecolor='firebrick', markeredgewidth=10)
```

```python
files = ['decimal_counting2.csv', 'revere.csv', 'test1.csv', 'test2.csv', 'test3.csv', 'test4.csv']
arr = []
gains = []
for f in files:
    df = pd.read_csv(f, on_bad_lines='skip', names=np.arange(0, 101, 1))
    av_column = df.mean(axis=0)
    arr.append(av_column)
    gain = round((av_column[100] - 10000) / 100, 2)
    gains.append(gain)
#print(arr)
df = pd.DataFrame({
    f'Card Removal Effect Method: {gains[0]}%': arr[0],
    f'Revere RAPC: {gains[1]}%': arr[1],
    f'More Practical Method: {gains[2]}%': arr[2],
   })
lines = df.plot.line()

lines.set_xlabel("Hands Played")
lines.set_ylabel("Player Bankroll")
```

```python
data = pd.read_csv("all.csv")
data['avg_profit'] = round((data['mean_bankroll']-10000)/100,3)
data.sort_values(by=['avg_profit'], ascending=False)
```

```python
new_data = pd.read_csv('test.csv')
new_data2 = pd.read_csv('practical.csv')
new_data['avg_profit'] = round((new_data['mean_bankroll']-10000)/100,3)
new_data2['avg_profit'] = round((new_data2['mean_bankroll']-10000)/100,3)
```

```python
hi_opt = data[data['card_counting']=="ZEN"]['avg_profit'].reset_index(drop=True)
revere = data[data['card_counting']=="Revere RAPC"]['avg_profit'].reset_index(drop=True)
decimal = new_data['avg_profit'].reset_index(drop=True)
practical = new_data2['avg_profit'].reset_index(drop=True)

frame = {'Hi Opt': hi_opt, 'Revere RAPC':revere, 'New Effect of Removal Method': decimal, 'Practical New Method': practical}
counting_methods = pd.DataFrame(frame)
```

```python
boxes = [hi_opt, revere, decimal, practical]
lists = []
for i in boxes:
    lists.append(list(i))
fig, ax = plt.subplots()
bp = ax.boxplot(x=lists, showmeans=True, boxprops=boxprops, medianprops=medianprops, flierprops=flierprops, whiskerprops=whiskerprops,
                capprops=capprops, meanprops=meanpointprops)
m1 = counting_methods.mean(axis=0)
st1 = counting_methods.std(axis=0)

for i, line in enumerate(bp['medians']):
    x, y = line.get_xydata()[1]
    text = " μ={:.2f}\n σ={:.2f}".format(m1[i], st1[i])
    ax.annotate(text, xy=(x, y), fontsize=25)
ax.set_xticks([1, 2, 3, 4], ['Hi-Opt II', 'Revere', 'Effect of Removal', 'Practical New Method'])
ax.set_ylabel("Average Profit")
ax.set_xlabel("Card Counting Method")
```
